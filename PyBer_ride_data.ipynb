```python
#Import dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statistics
```

```python
#Read the csv
pyber_ride_df = pd.read_csv("Resources/Pyber_ride_data.csv")
```

```python
#Additional plot details
y_axis = pyber_ride_df["Avg. Fare ($USD)"]
stdev = statistics.stdev(y_axis)

#Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=3)

#Set x-axis and tick locations
x_axis = np.arange(len(pyber_ride_df))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, pyber_ride_df["Month"])



plt.show()
```

```python
#Additional plot details
y_axis = pyber_ride_df["Avg. Fare ($USD)"]
stdev = statistics.stdev(y_axis)

#Plot the data
pyber_ride_df.plot(x="Month", y="Avg. Fare ($USD)", yerr=stdev, capsize=3, kind="bar", color="lightblue")

#Set x-axis and tick locations
x_axis = np.arange(len(pyber_ride_df))
plt.yticks(np.arange(0, 51, step=5.0))
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, pyber_ride_df["Month"], rotation=0)

plt.show()
```
